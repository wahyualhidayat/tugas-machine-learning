```python
import pandas as pd
liver = pd.read_csv('/content/liver.csv')
```

```python
liver.drop('Unnamed: 0',axis=1, inplace=True)
```

```python
liver.head (10)
```

```python
#memisakhan atribut dengan label dimana x itu atribut dan y itu label
X = liver[['Age','Sex','ALB','ALP','ALT','AST','BIL','CHE','CHOL','CREA','GGT','PROT']]
y = liver['Category']

# Label encoding untuk fitur 'Sex'
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
X['Sex'] = label_encoder.fit_transform(X['Sex'])

#membagi dataset menjadi data latih dan data uji
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.1, random_state=12)
```

```python
# Pre-processing: Mengganti nilai NaN dengan nilai rata-rata dari setiap kolom
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')
X_train = pd.DataFrame(imputer.fit_transform(X_train), columns=X_train.columns)
X_test = pd.DataFrame(imputer.transform(X_test), columns=X_test.columns)
```

```python
from sklearn.tree import DecisionTreeClassifier

#membuat model decision tree
tree_model= DecisionTreeClassifier()

#melatih model menggunakan data latih dengan fungsi fit
tree_model= tree_model.fit(X_train,y_train)
```

```python
#evaluasi model untuk meguji seberapa baik model yang digunakan dan akurasi

from sklearn.metrics import accuracy_score

y_pred =  tree_model.predict(X_test)

accuracy_score =round(accuracy_score(y_pred, y_test), 3)

print('Accuracy: ',accuracy_score*100,'%')
```
